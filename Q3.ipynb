```python
import cv2
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/8594461.png')
image2 = cv2.imread('/content/8624553.png')
image = cv2.resize(image, (224, 224))
image2 = cv2.resize(image2, (224, 224))
cv2_imshow(image)
cv2_imshow(image2)
```

```python
blended=cv2.add(image,image2)
cv2_imshow(blended)
```

```python
blended2=cv2.subtract(image,image2)
cv2_imshow(blended2)
```

```python
blended3=cv2.multiply(image,image2)
cv2_imshow(blended3)
```

```python
blended4=cv2.divide(image,image2)
cv2_imshow(blended4)
```
